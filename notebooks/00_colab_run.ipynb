## Set `results_dir` to Drive path

Use the next cell to rewrite `config/experiment.yaml` so outputs are persisted to Drive.  
Example path: `/content/drive/MyDrive/comp2024_results`

```python
from pathlib import Path
import yaml

cfg_path = Path('config/experiment.yaml')
cfg = yaml.safe_load(cfg_path.read_text(encoding='utf-8'))
cfg['output']['results_dir'] = '/content/drive/MyDrive/comp2024_results'
cfg_path.write_text(yaml.safe_dump(cfg, sort_keys=False), encoding='utf-8')
print('Updated results_dir ->', cfg['output']['results_dir'])
```
